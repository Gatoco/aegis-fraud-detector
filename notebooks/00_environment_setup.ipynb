## 5. Git Configuration and Verification

Verificamos la configuraci칩n de Git y el archivo .gitignore.

```python
import subprocess

# Verificar si es un repositorio Git
try:
    result = subprocess.run(['git', 'status'], 
                          capture_output=True, text=True, cwd=project_root)
    if result.returncode == 0:
        print("九 Repositorio Git inicializado")
        
        # Verificar remote
        remote_result = subprocess.run(['git', 'remote', '-v'], 
                                     capture_output=True, text=True, cwd=project_root)
        if remote_result.stdout:
            print("九 Remote configurado:")
            print(f"   {remote_result.stdout.strip()}")
        else:
            print("丘멆잺  No se encontr칩 remote configurado")
    else:
        print("仇 No es un repositorio Git")
except FileNotFoundError:
    print("仇 Git no est치 instalado o no est치 en PATH")
```

```python
# Verificar .gitignore
gitignore_path = project_root / '.gitignore'

if gitignore_path.exists():
    print("九 .gitignore encontrado")
    
    # Verificar que contenga reglas importantes
    with open(gitignore_path, 'r') as f:
        content = f.read()
    
    important_rules = [
        '__pycache__',
        '*.pyc',
        'venv/',
        '.env',
        'data/01_raw',
        'models/*.pkl'
    ]
    
    print("\n游늶 Verificando reglas importantes en .gitignore:")
    for rule in important_rules:
        status = "九" if rule in content else "仇"
        print(f"   {rule}: {status}")
else:
    print("仇 .gitignore no encontrado")
    print("   Debe crearse para evitar commitear archivos innecesarios")
```
